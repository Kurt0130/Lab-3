```python
import numpy as np
import matplotlib.pyplot as plt
def f(t):
    y = np.exp(-t/2)
    return y

X = np.linspace(0,10,1000)
plt.plot(X, f(X))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
T = np.arange(0, 8)
Y = np.array([0, 1e-3, 1e-2, 1e-1, 1, 1e1, 1e2, 1e3])


if len(T) != len(Y):
    raise ValueError("T and Y must be of the same length for plotting")

plt.plot(T, Y, marker='o')  
plt.xlabel('T')
plt.ylabel('Y')
plt.title('Plot of Y vs. T')
plt.yscale('log')  
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
f = np.sin
N_low = np.linspace(0, 2 * np.pi, 100)  # Use more points for a smoother plot


plt.plot(N_low, f(N_low), color='blue', label='Sine function')


plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of sin(x)')
plt.legend()
plt.grid(True)


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
f = np.sin
N_high = np.linspace(0, 2 * np.pi, 100)  

plt.figure(figsize=(12, 6))  
plt.plot(N_high, f(N_high), color='blue', label='Sine function')
plt.stem(N_high, f(N_high), linefmt='--', markerfmt='o', basefmt='r-', label='Discrete points')


plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of sin(x) with Discrete Points')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
f = np.sin
N = np.linspace(0, 2 * np.pi, int(4 * np.pi))  


plt.figure(figsize=(12, 6))  
plt.plot(N, f(N), color='blue', label='Continuous sine function')
plt.stem(N, f(N), linefmt='--', markerfmt='o', basefmt='r-', label='Discrete points')


plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of sin(x) with Continuous and Discrete Points')
plt.legend()
plt.grid(True)


plt.show()
```
